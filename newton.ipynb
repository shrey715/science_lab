```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
dval = np.array([])# enter Vn values
dval
```

```python
xval = [180*i for i in range(1,8)]
xval
```

```python
plotx = xval
ploty = np.log(dval/dval[0])
print(plotx,ploty)
```

```python
(m,c) = np.linalg.lstsq(np.vstack([plotx, np.ones(len(plotx))]).T, ploty, rcond=None)[0]
print(m,c)
```

```python
plt.figure(figsize = (8,11))
plt.plot(np.linspace(min(plotx),max(plotx),10),m*np.linspace(min(plotx),max(plotx),10)+c)
plt.scatter(plotx,ploty,c = 'r')
plt.title(label = "D^2 vs Ring number")
plt.xlabel("Ring number")
plt.ylabel("D^2")
for xy in zip(plotx, ploty):
   plt.annotate('(%.1f, %.7f)' % xy, xy=xy)
plt.grid(True,color='g')
plt.savefig("trial3.png")
```
